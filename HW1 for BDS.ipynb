#### Preprocessing

##### 1. Handing Categorical Data

###### Reason:  All data is categorical data which is difficult for us to deal with. So we need to turn them to numeral data.

```python
a_mapping = {'low': 0, 'mid': 1, 'high': 2}
df['L-CORE'] = df['L-CORE'].map(a_mapping)
b_mapping = {'low': 0, 'mid': 1, 'high': 2}
df['L-SURF'] = df['L-SURF'].map(b_mapping)
c_mapping = {'poor': 0, 'fair': 1, 'good': 2, 'excellent': 3}
df['L_O2'] = df['L_O2'].map(c_mapping)
d_mapping = {'low': 0, 'mid': 1, 'high': 2}
df['L-BP'] = df['L-BP'].map(d_mapping)
e_mapping = {'unstable': 0, 'mod-stable': 1, 'stable': 2}
df['SURF-STBL'] = df['SURF-STBL'].map(e_mapping)
f_mapping = {'unstable': 0, 'mod-stable': 1, 'stable': 2}
df['CORE-STBL'] = df['CORE-STBL'].map(f_mapping)
g_mapping = {'unstable': 0, 'mod-stable': 1, 'stable': 2}
df['BP-STBL'] = df['BP-STBL'].map(g_mapping)
h_mapping = {'I': 0, 'S': 1, 'A': 2, 'A ':2}
df['ADM-DECS'] = df['ADM-DECS'].map(h_mapping)

```

###### Reason: data "ADM-DECS" is label, so it must be changed.

```python
from sklearn.preprocessing import LabelEncoder

class_le = LabelEncoder()
y = class_le.fit_transform(df['ADM-DECS'].values)
y
```
